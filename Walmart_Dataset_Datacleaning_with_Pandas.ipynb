```python
# Import Pandas Dataframe
import pandas as pd

df = pd.read_csv('/content/Walmart.csv')
```

```python
# Print the number of rows and columns in the dataset
df.shape
```

```python
# Prints the statistics of the dataset
df.describe()
```

```python
# Prints the columns with data_type information
df.info()
```

```python
# prints the sum of duplicate rows in the dataset

df.duplicated().sum()
```

```python
# prints the sum of null values in the dataset

df.isnull().sum()
```

```python
# Drops the duplicates and prints the shape of the dataset
df.drop_duplicates(inplace = True)
df.duplicated().sum()
df.shape
```

```python
# Drops the null values and prints the sum of null values based on the columns
df.dropna(inplace = True)
df.isnull().sum()
```

```python
# change the datatye of unit_price from object to float
df.dtypes
```

```python
# Replacing the $ sign with ' ' before converting unit_price from object to float
df['unit_price'] = df['unit_price'].str.replace('$','').astype(float)
```

```python
df.info()
```

```python
# prints the column names in the dataset
df.columns
```

```python
#creating the total column
df['total'] = df['unit_price'] * df['quantity']
```
